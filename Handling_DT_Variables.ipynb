```python
date['date_year'] = date['date'].dt.year
date['date_month'] = date['date'].dt.month
date['date_day'] = date['date'].dt.day
date['date_dayofweek'] = date['date'].dt.dayofweek
date['date_dayofyear'] = date['date'].dt.dayofyear
date['date_weekofyear'] = date['date'].dt.isocalendar().week
```

```python
date['date_month_name'] = date['date'].dt.month_name()
```

```python
date.head()
```

```python
date['date_day_name'] = date['date'].dt.day_name()
```

```python
date['date_is_weekend'] = np.where(date['date_day_name'].isin(['Saturday','Sunday']),1,0)
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
date['days_difference'] = (today - date['date']).dt.days
```

```python
date['months_difference'] = ((today.year - date['date'].dt.year) * 12 +
                              (today.month - date['date'].dt.month))
```
